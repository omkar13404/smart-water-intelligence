```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# -----------------------------
# CONFIG
# -----------------------------

NUM_ROWS = 25000

# -----------------------------
# STORAGE
# -----------------------------

data = []

# -----------------------------
# DATE GENERATION
# -----------------------------

start_date = datetime(2024, 1, 1)

# -----------------------------
# GENERATE DATA
# -----------------------------

for i in range(NUM_ROWS):

    date = start_date + timedelta(hours=i)

    month = date.month

    # -----------------------------
    # SEASON LOGIC
    # -----------------------------

    if month in [6, 7, 8, 9]:
        season = "Monsoon"
        rainfall = random.randint(20, 120)
        humidity = random.randint(70, 95)
        temperature = random.randint(22, 32)

    elif month in [3, 4, 5]:
        season = "Summer"
        rainfall = random.randint(0, 20)
        humidity = random.randint(35, 65)
        temperature = random.randint(30, 42)

    else:
        season = "Winter"
        rainfall = random.randint(0, 40)
        humidity = random.randint(40, 75)
        temperature = random.randint(15, 28)

    # -----------------------------
    # WATER STORAGE
    # -----------------------------

    tank_capacity = random.choice([1000, 1500, 2000])

    water_level = random.randint(100, tank_capacity)

    daily_usage = random.randint(80, 350)

    occupancy = random.randint(2, 12)

    # -----------------------------
    # WATER COLLECTION LOGIC
    # -----------------------------

    roof_area = random.randint(50, 200)

    efficiency = round(random.uniform(0.5, 0.95), 2)

    water_collected = (
        rainfall * roof_area * efficiency
    ) / 10

    # -----------------------------
    # OVERFLOW LOGIC
    # -----------------------------

    future_level = water_level + water_collected - daily_usage

    if future_level > tank_capacity:
        overflow_risk = 1
    else:
        overflow_risk = 0

    # -----------------------------
    # ANOMALY / LEAKAGE
    # -----------------------------

    leakage = 0

    if random.random() < 0.05:
        leakage = 1
        daily_usage += random.randint(150, 300)

    # -----------------------------
    # SUSTAINABILITY SCORE
    # -----------------------------

    sustainability_score = (
        100
        - (daily_usage / 5)
        + (water_collected / 50)
    )

    sustainability_score = max(
        0,
        min(100, sustainability_score)
    )

    # -----------------------------
    # APPEND ROW
    # -----------------------------

    data.append([
        date,
        season,
        rainfall,
        humidity,
        temperature,
        tank_capacity,
        water_level,
        daily_usage,
        occupancy,
        roof_area,
        efficiency,
        round(water_collected, 2),
        overflow_risk,
        leakage,
        round(sustainability_score, 2)
    ])

# -----------------------------
# CREATE DATAFRAME
# -----------------------------

columns = [
    "datetime",
    "season",
    "rainfall_mm",
    "humidity",
    "temperature",
    "tank_capacity",
    "water_level",
    "daily_usage",
    "occupancy",
    "roof_area",
    "collection_efficiency",
    "water_collected",
    "overflow_risk",
    "leakage_detected",
    "sustainability_score"
]

df = pd.DataFrame(data, columns=columns)

# -----------------------------
# SAVE CSV
# -----------------------------

df.to_csv(
    "data/smart_water_dataset.csv",
    index=False
)

print("\nDataset Generated Successfully!")
print(df.head())
print("\nShape:", df.shape)
```
