```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# -----------------------------
# LOAD FORECAST
# -----------------------------

forecast = pd.read_csv(
    "../../data/forecast_results.csv"
)

# -----------------------------
# CONVERT DATE
# -----------------------------

forecast["ds"] = pd.to_datetime(
    forecast["ds"]
)

# -----------------------------
# PLOT
# -----------------------------

fig, ax = plt.subplots(
    figsize=(14, 6)
)

ax.plot(
    forecast["ds"],
    forecast["yhat"]
)

# -----------------------------
# FORMAT DATES
# -----------------------------

ax.xaxis.set_major_locator(
    mdates.MonthLocator()
)

ax.xaxis.set_major_formatter(
    mdates.DateFormatter('%b %Y')
)

plt.xticks(rotation=45)

# -----------------------------
# TITLES
# -----------------------------

plt.title(
    "Predicted Future Water Collection"
)

plt.xlabel("Date")

plt.ylabel("Forecasted Collection")

plt.tight_layout()

plt.show()
```
