```python
import pickle
import numpy as np

# -----------------------------
# LOAD MODEL
# -----------------------------

model = pickle.load(
    open(
        "../../models/anomaly_model.pkl",
        "rb"
    )
)

# -----------------------------
# SAMPLE INPUT
# -----------------------------

sample = np.array([
    [
        0,      # season
        5,      # rainfall
        60,     # humidity
        30,     # temperature
        200,    # water level
        700,    # daily usage (abnormal)
        2,      # occupancy
        20      # water collected
    ]
])

# -----------------------------
# PREDICT
# -----------------------------

prediction = model.predict(sample)

# -----------------------------
# OUTPUT
# -----------------------------

if prediction[0] == -1:

    print(
        "\nWARNING: Possible Leakage / Anomaly Detected!"
    )

else:

    print(
        "\nSystem Operating Normally"
    )
```
