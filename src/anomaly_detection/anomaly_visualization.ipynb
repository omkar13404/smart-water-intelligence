```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# LOAD RESULTS
# -----------------------------

df = pd.read_csv(
    "../../data/anomaly_results.csv"
)

# -----------------------------
# NORMAL DATA
# -----------------------------

normal = df[
    df["anomaly_prediction"] == 0
]

# -----------------------------
# ANOMALIES
# -----------------------------

anomalies = df[
    df["anomaly_prediction"] == 1
]

# -----------------------------
# PLOT
# -----------------------------

plt.figure(figsize=(12, 6))

plt.scatter(
    normal["water_level"],
    normal["daily_usage"],
    label="Normal"
)

plt.scatter(
    anomalies["water_level"],
    anomalies["daily_usage"],
    label="Anomaly"
)

plt.xlabel("Water Level")

plt.ylabel("Daily Usage")

plt.title(
    "Anomaly Detection Visualization"
)

plt.legend()

plt.show()
```
