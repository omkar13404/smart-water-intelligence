```python
import pickle
import numpy as np

# -----------------------------
# LOAD MODEL
# -----------------------------

model = pickle.load(
    open(
        "../../models/collection_model.pkl",
        "rb"
    )
)

# -----------------------------
# SAMPLE INPUT
# -----------------------------

sample = np.array([
    [
        0,      # season
        80,     # rainfall
        90,     # humidity
        28,     # temperature
        1500,   # tank capacity
        600,    # water level
        120,    # daily usage
        5,      # occupancy
        120,    # roof area
        0.85    # efficiency
    ]
])

# -----------------------------
# PREDICT
# -----------------------------

prediction = model.predict(sample)

print(
    "\nPredicted Water Collection:"
)

print(
    f"{prediction[0]:.2f} liters"
)
```
