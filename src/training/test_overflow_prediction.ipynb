```python
import pickle
import numpy as np

# -----------------------------
# LOAD MODEL
# -----------------------------

model = pickle.load(
    open(
        "../../models/overflow_model.pkl",
        "rb"
    )
)

# -----------------------------
# SAMPLE INPUT
# -----------------------------

sample = np.array([
    [
        0,      # season
        100,    # rainfall
        92,     # humidity
        27,     # temperature
        1000,   # tank capacity
        950,    # current water level
        80,     # daily usage
        6,      # occupancy
        150,    # roof area
        0.90    # efficiency
    ]
])

# -----------------------------
# PREDICTION
# -----------------------------

prediction = model.predict(sample)

probability = model.predict_proba(sample)

# -----------------------------
# OUTPUT
# -----------------------------

if prediction[0] == 1:

    print("\nWARNING: Overflow Risk!")

else:

    print("\nTank Status: Safe")

print(
    f"\nOverflow Probability: "
    f"{probability[0][1] * 100:.2f}%"
)
```
